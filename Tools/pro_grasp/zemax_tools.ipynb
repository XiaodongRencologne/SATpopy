```python
import numpy as np
import matplotlib.pyplot as plt
Lambda_0 = 3.33102731
```

```python
def FT_far(x,y,Field,Lambda):
    X = x/Lambda
    Y = y/Lambda
    dx = np.abs(X[0]-X[1])
    dy = np.abs(Y[0]-Y[1])
    theta_x = np.fft.fftshift(np.fft.fftfreq(x.size, d=dx))
    theta_y = np.fft.fftshift(np.fft.fftfreq(y.size, d=dy))
    beam = np.fft.ifftshift(np.fft.fft2(np.fft.fftshift(Field)))
    return theta_x, theta_y, beam
```

```python
AP = np.genfromtxt('zemax_data/90GHz_center_amp0.txt',skip_header = 16)
AP_phase = np.genfromtxt('zemax_data/90GHz_center_phase0.txt',skip_header = 16)
AP_field= AP*np.exp(1j*AP_phase)
```

```python
x0 = np.linspace(-2048/2*6.19189453125,(2048/2-1)*6.19189453125,2048)
```

```python
tx,ty,BEAM = FT_far(x0,x0,AP_field,Lambda_0)
```

```python
n=5
for item in range(-3,3+1):
    plt.plot(tx[600:-600]*180/np.pi,np.log10(np.abs(BEAM[1024-item*n,600:-600]))*20)
plt.show()
for item in range(-3,3+1):
    plt.plot(tx[600:-600]*180/np.pi,np.angle(BEAM[1024-item*n,600:-600]))
plt.show()
```
